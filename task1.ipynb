```python
import pandas as pd

sentiment = pd.read_csv("D:\\dstask\\historical_data.csv")
trades = pd.read_csv("D:\\dstask\\historical_data.csv")

sentiment["Date"] = pd.to_datetime(sentiment["Date"]).dt.date

trades["time"] = pd.to_datetime(trades["time"])
trades["Date"] = trades["time"].dt.date

df = trades.merge(
    sentiment,
    on="Date",
    how="left"
)

print(df.head())
print(df.shape)
```
